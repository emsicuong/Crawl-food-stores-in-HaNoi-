```python
cities_configs = func_get_congigs_cities()

for city in cities_configs:
    print("City ID: ", city['city']['id'])
    print("City Name: ", city['city']['name'])
    print("City slug: ", city['city']['slug'])
    cates = func_get_cates_from_cityid(cityId=city['city']['id'])
    
    if(city['city']['id'] == 1):
        #Các ngành hàng trong 1 thành phố
        for cate in cates:
            print(cate['id'])
            
            #Tính số trang của 1 ngành hàng
            _, data_pages = get_shop_info_summary(categoriesID=cate['id'], cityId=city['city']['id'], 
                                                page=1, limit=24)
            pages = data_pages['total']//data_pages['limit'] + 1
            
            #Các trang trong 1 ngành hàng
            for num_page in range(0, pages):
                print(f"Trang số {num_page}")
                data_shops, _ = get_shop_info_summary(categoriesID=cate['id'], cityId=city['city']['id'],
                                                page=num_page, limit=24)
                document_shops = []
                #Các quán trong 1 trang
                for i, shop in enumerate(data_shops):
                    print(i, "slug: ", shop['slug'])
                    shop_info = get_shop_full_info(slug=shop['slug'])
                    menu_info = get_menu_info(slug=shop['slug'])

                    doc_shop = {
                        'city': city['city'],
                        'category': cate,
                        'shop_info': shop_info,
                        'menu_info': menu_info,
                        'createAt': time.time()
                    }
                    
                    document_shops.append(doc_shop)
                    
                #Kiểm tra nếu có dữ liệu thì mới insert
                if len(document_shops) > 0:
                    x = mycol.insert_many(document_shops)
                    #print(x.inserted_ids)
                    print(f"Lưu trữ dữ liệu cửa hàng {cate['value']} tại {city['city']['name']} thành công")
        break
```
